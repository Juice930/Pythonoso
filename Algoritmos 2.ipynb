```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mincuad(x,y):
    x,y=np.array(x),np.array(y)
    m=float(len(x)*np.sum(x*y)-np.sum(x)*np.sum(y))/(len(x)*np.sum(x**2)-np.sum(x)**2)
    b=np.mean(y)-m*np.mean(x)
    R2=1-np.sum((y-(m*x+b))**2)/np.sum((y-np.mean(y))**2)
    return m,b,R2
```

```python
print mincuad([1,2,3],[1,2,2])
```

# Derivación Numérica

```python
def f(x):
    return 3*x**3+2*x**2+5*x+1
x=np.arange(-2,2,0.001)
plt.plot(x,f(x))
plt.show()
```

## Derivada hacia Adelante

```python
x=np.arange(-2,2,0.001)
y=f(x)
dy=[(y[i+1]-y[i])/(x[i+1]-x[i]) for i in range(0,len(x)-1)]
plt.plot(x[:-1],dy)
plt.show()
```

## Derivada hacia atrás

```python
x=np.arange(-2,2,0.001)
y=f(x)
dy=[(y[i]-y[i-1])/(x[i]-x[i-1]) for i in range(1,len(x))]
plt.plot(x[1:],dy)
plt.show()
```

## Derivada Central

```python
x=np.arange(-2,2,0.001)
y=f(x)
dy=[(y[i+1]-y[i-1])/(2*(x[i+1]-x[i])) for i in range(1,len(x)-1)]
plt.plot(x[1:-1],dy)
plt.show()
```

```python
def fp(x):
    return 9*x**2+3*x+5
def fpp(x):
    return 18*x+3
```

```python
x=np.arange(-2,2,0.01)
y=f(x)
dy=[(y[i+1]-y[i-1])/(2*(x[i+1]-x[i])) for i in range(1,len(x)-1)]
plt.plot(x[1:-1],dy,x,fp(x),'r')
plt.show()
```

```python
x=np.arange(-2,2,0.01)
dy=[(y[i+1]-y[i-1])/(2*(x[i+1]-x[i])) for i in range(1,len(x)-1)]
ddy=[(dy[i+1]-dy[i-1])/(2*(x[i+1]-x[i])) for i in range(1,len(dy)-1)]
plt.plot(x[2:-2],ddy,x,fpp(x),'r')
plt.show()
```
